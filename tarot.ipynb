```python
import sqlite3
import random
from datetime import datetime
import time
import os
```

```python
os.makedirs("data", exist_ok=True)
conect= sqlite3.connect("data/tarot.db")
cursor = conect.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS cartas(
               id INTEGER PRIMARY KEY AUTOINCREMENT,
               nome TEXT NOT NULL,
               posicao TEXT NOT NULL,
               significado TEXT NOT NULL,
               conselho TEXT NOT NULL             
)
 """)
conect.commit()
```

```python
def cadastrar_carta(nome, significado_normal, conselho_normal, significado_invertido, conselho_invertido):
    cursor.execute("SELECT id FROM cartas WHERE nome = ?", (nome,))
    if cursor.fetchone()is None:
        cursor.execute("""
            INSERT INTO cartas(nome, posicao, significado, conselho)
            VALUES (?, 'normal', ?,?)
         """,(nome,significado_normal,conselho_normal))

        cursor.execute("""
            INSERT INTO cartas(nome, posicao, significado, conselho)
            VALUES (?, 'invertida', ?,?)
         """,(nome,significado_invertido,conselho_invertido))

        conect.commit()
```

```python
def sorteio_cartas():
    print('___________________________________________________________________________________________________________________________\n')
    print(f"🔮 - Concentre-se: Pense claramente em sua pergunta...")
    time.sleep(10.0)
    print(f"🃏 - Embaralhamento: Os arcanos estão sendo embaralhados...")
    time.sleep(1.3)
    print(f'✨ - Tiragem: A sua carta será revelada...')
    time.sleep(1.0)

def conselho_tarot_sql():
    posicao_sortida = 'invertida' if random.random() < 0.25 else 'normal'

    cursor.execute("""
    SELECT nome, posicao, significado, conselho
    FROM cartas
    WHERE posicao = ?
    ORDER BY RANDOM()
    LIMIT 1
""",(posicao_sortida,))
    
    carta = cursor.fetchone()
    
    if carta:
        nome, posicao, significado, conselho = carta
        print('___________________________________________________________________________________________________________________________\n')
        print(f' 🔮 - CONSELHO DO DIA: {nome}. Posição:({posicao.upper()})')
        print(f' 💫 - Significado: {significado}')
        print(f' 🌙 - Conselho: {conselho}')
        print('___________________________________________________________________________________________________________________________')
    else:
        print('Nenhuma carta cadastrada no banco de dados.')
    
```

```python
sorteio_cartas()
conselho_tarot_sql()
conect.close()
```
